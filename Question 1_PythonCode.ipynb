#### 2. Normality Test

##### 2.1 Q-Q Plot

```python
# Simulating data with the same structure as the user's dataset
np.random.seed(0)

# Function to create Q-Q plots
def qq_plots(data):
    fig, axs = plt.subplots(3, 3, figsize=(15, 15))
    axs = axs.flatten()

    for i, col in enumerate(data.columns):
        stats.probplot(data[col], dist="norm", plot=axs[i])
        axs[i].set_title(f'Q-Q Plot for variable {col}')

    plt.tight_layout()
    plt.show()

# Calling the function to create Q-Q plots
qq_plots(data)
```

##### 2.2 Shapiro Wilk Test

```python
# Shapiro-Wilk Test for normality
normality_results = {column: shapiro(data[column]).pvalue for column in data.columns}
normality_results
```

#### 4. Kruskal-Wallis H Test

```python

# Reshape the dataframe to long format:
df_long = pd.melt(data, var_name='Group', value_name='Value')

# Perform the Kruskal-Wallis H Test
# Group the data by 'Group' and then apply the test to the groups.
h, p = stats.kruskal(*[group['Value'].values for name, group in df_long.groupby('Group')])

# Print the results
print('H-statistic:', h)
print('P-value:', p)

# Interpret the result
alpha = 0.05  # Or your chosen alpha level
if p < alpha:
    print("Reject the null hypothesis - there is a significant difference between the groups.")
else:
    print("Fail to reject the null hypothesis - we cannot conclude that there is a significant difference between the groups.")
```

#### 5. Clustering

```python
# Standardize the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
```

##### 5.1 Elbow Method

```python
# Determine the optimal number of clusters using the elbow method
inertia = []
k_range = range(1,11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state= 42)
    kmeans.fit(data_scaled)
    inertia.append(kmeans.inertia_)

# Plotting the Elbow Curve
plt.figure(figsize=(8,6))
plt.plot(k_range, inertia, marker = 'o')
plt.title('Elbow Method to Determine Optimal K')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

##### 5.2 Result with Silhoette Score

```python
# Run K-Means for 3 clusters
kmeans = KMeans(n_clusters=3, init='k-means++', random_state= 42)
cluster_labels = kmeans.fit_predict(data_scaled)

# Add the cluster labels back to our original dataframe for 3 clusters
data['Cluster'] = cluster_labels


# Evaluate the Clustering for 3 clusters
silhouette_avg = silhouette_score(data_scaled, cluster_labels)
print(f'Silhouette Score for 3 clusters: {silhouette_avg:.2f}')


# Run K-Means for 4 clusters
kmeans_4 = KMeans(n_clusters=4, init='k-means++', random_state=42)
cluster_labels_4 = kmeans_4.fit_predict(data_scaled)

# Add the cluster labels back to our original dataframe for 4 clusters
data['Cluster_4'] = cluster_labels_4

# Evaluate the Clustering for 4 clusters
silhouette_avg_4 = silhouette_score(data_scaled, cluster_labels_4)
print(f'Silhouette Score for 4 clusters: {silhouette_avg_4:.2f}')
```

```python
data = data.drop('Cluster_4', axis= 1)
```
